```python
'''
2. Multiple Linear Regression
    2.1 Import the Libraries
    2.2 Import the Dataset
    2.3 Encoding categorical data
    2.4 Splitting the dataset into the Training set and Test set
    2.5 Training the Simple Linear Regression model on the Training set
    2.6 Predicting the Test set results
--------------------------------------------------------------------------------
y^ = b0 + b1X1 + b2X2 + b3X3 + ......  + bnXn

y^ is the dependent variaable
b0 is the y intercept

b1 slope coeff1
X1 Independent variable1
b2 slope coeff2
X2 Independent variable2
b3 slope coeff3
X3 Independent variable3...


There is no need to aply FEATURE SCALING as the coeff compensate the variables
--------------------------------------------------------------------------------
Assumptions in Linear Regression:

1. Linearity
2. Homoscedasticity (Equal Variance- no increasin/decreasing cone be visible)
3. Multivariate Normality (Normality of Equal Distribution)
4. Independence of Observations (no pattern should be present)
5. Lack of Multicollinearity (predictorss not to be corelated to each other)
6. The Outllier Check

No need to check these in Multiple Regression
--------------------------------------------------------------------------------
Dummy Variables :

Profit | R&D spent | Admin | Marketing | State
    $       $          $        $          ?- Categorical Value(specefic states)

To handle categorical values, we make use of Dummy Variables, Each state is then
converted to its stack like table Present_state->1 Absent_state->0
just like light switches on and off

Dummy Variable Trap: can have only 1 dummy variable, if we add more than one

Conclusion: When making a model, omit one dummy variable, i.e if you have 4,
then have 3 (n-1)

remember One Hot Encoding, similar related concept
--------------------------------------------------------------------------------
Statistical Significance

Lets say we Toss a coin in World 1 -> Fair World 2-> Unfair

Tails 0.5
Tails 0.25
Tails 0.12
Tails 0.06 ----- below is less than 5%, where one qusestions Significance
Tails 0.03
Tails 0.01     (as we go down statistical significance and P value decreases)

Only in Fair world on can question. in unfair world, no need to Question
--------------------------------------------------------------------------------

5 Methods of building models:

1. All -in
2. Backward Elimination  ------->|
3. Forward Elimination           |--> Stepwise Regression
4. Bi-Directional Elimination--->|
5. Score Comparison

1. All -in : Throw in all the variables, Prior Knowledge with no other option,
             or when preparing for Backward Elimination

2. Backward : Step 1> Select a significance level to STAY in the model(SL =0.05)
              Step 2> Fit the full model with all possible predictors
              Step 3> Consider the predictor with HIGHEST P value. If P>SL, goto
                      step 4, else goto FIN
              Step 4> Remove the Predictor
              Step 5> Fit the model without this variable, i.e rebuild again
              Step 6> Go to Step 3 till the condition P>SL fails
              FIN   > Model Ready

3. Forward : Step 1> Select a significance level to ENTER the model(SL =0.05)
             Step 2> Fit all simple regression models y~xn, Select the one with
                     the lowest P-value
             Step 3> Keep this variable and fit all possible models with one
                     extra predictor added to the one(s) you already have
             Step 4> Consider the predictor with the lowest P-value. If P<SL,
                     goto Step 3, else goto FIN
             FIN   > Keep the pprevious model (as you just added a variable
                     with low SL, XD )

4. Bi-Directional : Step 1> Select a significance level to ENTER and to STAY
                            in the model (e.g. SLstay=0.05, SLenter = 0.05)
                    Step 2> Perform the next step of Forward selection (NEW
                            variables must have: P< SLenter)
                    Step 3> Perform Backward Elemination (OLD variables must
                            have P<SLstay to stay)
                    Step 4> Go to Step 2, until no NEW variables can enter or
                            no OLD variables can exit
                    FIN   > Model Ready

5. Score Comparison/ : Step 1> Select a Criterion of a good fit (e.g. Akaike )
   All Possible mod    Step 2> Construct ALL POSSIBLE reg models: (2^n - 1) mods
                       Step 3> Select the one with the best criterion
                       FIN   > Model Ready


Backward Elimination is irrelevant in Python, because the Scikit-Learn library
automatically takes care of selecting the statistically significant features
when training the model to make accurate predictions.
--------------------------------------------------------------------------------

'''
```

```python
# STEP 1: to import the required libraries

import numpy as np
import matplotlib.pyplot  as plt
import pandas as pd
```

```python
# STEP2 :is to import the required dataset

dataset = pd.read_csv('50_Startups.csv')
X =  dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Step 3 is for Encoding categorical data

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
# Step 4 is for Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Step 5 is for Training the Simple Linear Regression model on the Training set

'''
Q. Do we need to worry about the dummy variable trap?
ANS. No need to do anything to avoid Dummy variable trap as the below imported
class for MLR automatically avoids this trap

Q. Do we have to work on features to select the best ones with high Pvalues with
the most statistical significance?
ANS. The class called below will automatically identify the best features
'''

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Step 6 is for Predicting the Test set results

y_pred = regressor.predict(X_test)
np.set_printoptions(precision = 2)   # for any numerical value upto 2 decimals
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape
 (len(y_test),1)),1))
```

```python
# Making a single prediction
'''
 (for example the profit of a startup with
R&D Spend = 160000, Administration Spend = 130000, Marketing Spend = 300000
and State = 'California')
'''

print(regressor.predict([[1, 0, 0, 160000, 130000, 300000]]))
```
