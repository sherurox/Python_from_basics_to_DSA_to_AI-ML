```python
'''
5.1_Apriori
    3.1  Import the Libraries
    3.2  Data Preprocessing
    3.3  Training the Apriori model on the dataset
    3.4  Displaying the first results coming directly from the output of
         the apriori function
    3.5  Putting the results well organised into a Pandas DataFrame
    3.6  Displaying the results non sorted
    3.7  Displaying the results sorted by descending lifts
--------------------------------------------------------------------------------
People who bought this also bought that (Recomendation)

Support, Confidence, Lift = (Confidence / Support)

1> Set a minimum suport and confidence
2> Take all subsets in transactions having higher support than minimum support
3> Take all the rules of these subsets having higher confidence than minimum
   confidence
4> Sort all the rules by decreasing lift
--------------------------------------------------------------------------------
'''
```

```python
# Step1: Import the Libraries

!pip install apyori

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Step2: Data Preprocessing

dataset = pd.read_csv('Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
  transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
# Step3: Training the Apriori model on the dataset

from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, #(3*7)/7501
                min_confidence = 0.2, min_lift = 3, min_length = 2,
                max_length = 2)  #lift =3to5 length is buy 1 get 1
```

```python
# Step4: Displaying the first results coming directly from the output of
#the apriori function

results = list(rules)
```

```python
# Step5: Putting the results well organised into a Pandas DataFrame

def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results),
                                  columns = ['Left Hand Side', 'Right Hand Side'
                                  , 'Support', 'Confidence', 'Lift'])
```

```python
# Step6: Displaying the results non sorted

resultsinDataFrame
```

```python
# Step7: Displaying the results sorted by descending lifts

resultsinDataFrame.nlargest(n = 10, columns = 'Lift')
```
