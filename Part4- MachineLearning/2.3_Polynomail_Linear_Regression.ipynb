```python
'''
2.3 Polynomial Linear Regression
    2.1 Import the Libraries
    2.2 Import the Dataset
    2.3 Training the Linear Regression model on the whole dataset
    2.4 Training the Polynomial Regression model on the whole dataset
    2.5 Visualising the Linear Regression results
    2.6 Visualising the Polynomial Regression results
    2.7 Visualising the Polynomial Regression results (for higher resolution
                                                        and smoother curve)
    2.8 Predicting a new result with Linear Regression
    2.9 Predicting a new result with Polynomial Regression
--------------------------------------------------------------------------------
y^ = b0 + b1X1 + b2X1^2 + b3X1^3 + ......  + bnX1^n

y^ is the dependent variaable
b0 is the y intercept
b1   slope coeff1
X1   Independent variable1, same across all only powert increases
b2   slope coeff2
b3   slope coeff3

Why is it called Linear?
Ans. We talk about the coeff and not the independent variables
--------------------------------------------------------------------------------

'''
```

```python
# STEP 1: to import the required libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# STEP2 :is to import the required dataset

dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

```python
# STEP3 :is for Training the Linear Regression model on the whole dataset

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# STEP4 :is for Training the Polynomial Regression model on the whole dataset

from sklearn.preprocessing import PolynomialFeatures # we are pre processing
# the non linear variables for our equation in the model
#now we will be making a matrix of features
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)  #to transform the column feature (Level
# here in csv)  to matrix of feature
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
# STEP5 :is for Visualising the Linear Regression results

plt.scatter(X, y, color ='red')
plt.plot(X, lin_reg.predict(X), color ='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# STEP6 :is for Visualising the Polynomial Regression results

plt.scatter(X, y, color ='red')
plt.plot(X, lin_reg_2.predict(X_poly), color ='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
# STEP7 :is for Visualising the Polynomial Regression results (for higher
                                                #resolution and smoother curve)

X_grid = np.arange(min(X), max(X), 0.1) # increasing the density from 1,2,3 to
                                        #1.1,1.2,1.3...
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

```python
# STEP8 :is for Predicting a new result with Linear Regression

lin_reg.predict([[6.5]]) #sqr brackets means 2d matrix, i.e. rows n columns
```

```python
# STEP9 :is for Predicting a new result with Polynomial Regression

lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
