```python
# Now comes the popping

def pop(self):
    #case 1 is empty list
    if self.length==0:
        return None
    else: #case 2 the rest...
        temp=self.head
        pre=self.head
        while (temp.next):
            pre=temp
            temp=temp.next
        self.tail=pre
        self.tail.next=None
        self.length -= 1
        if self.length==0 : #case 3 if only one node present
            self.head=None
            self.tail=None
        return temp.value
```

```python
# Now we learn prepend

def prepend(self,value):
    new_node= Node(value)
    if self.length==0: #case 1 empty list
        self.head=new_node
        self.tail=new_node
    else:              # case 2 normal list
        new_node.next=self.head
        self.head=new_node
    self.length += 1 
    return True 
```

```python
# Now we learn to popfirst

def popfirst(self):
    if self.length==0: #case 1 If the list is empty
        return None
    else:
        temp=self.head #case 2 normal list
        self.head=self.head.next
        temp.next=None
        self.length -= 1
    if self.length==0:  # case 3 when we have only 1 node , similar to popping
        self.tail= None
    return temp.value
```

```python
# now we learn how to search 

def get(self,index):
    if index <0 or index>=self.length:  #case 1 input of incorrect index
        return None
    temp=self.head
    for _ in range(index):
        temp=temp.next
    return temp.value
```

```python
#now we learn how to set, just like updating a value of a node

def set_value(self,value,index):
    temp=self.get(index)
    if temp: #similar to to if temp is not none
        temp.value=value
        return True
    return False
```

```python
# Now we learn how to insert a node in a linked list

def insert(self,value,index):
      if index <0 or index>=self.length: # for index out of range
            return False
      if index == 0:
            return self.prepend(value) # for index 0 
      if index == self.length:
            return self.append(value)   # for index at last
      new_node = Node(value)
      temp = self.get(index-1)
      new_node.next= temp.next
      temp.next=new_node
      self.length +=1
      return True
```

```python
# Now we learn how to remove an item from linked list

def remove(self,index):
    if index <0 or index>=self.length: # for index out of range
            return None  # not false its None
    if index == 0:
            return self.pop_first() # for index 0 
    if index == self.length - 1:
            return self.pop()   # for index at last
    prev = self.get(index-1)
    temp= prev.next
    prev.next=temp.next
    temp.next=None
    self.length -=1
    return temp.value
```

```python
# revision 

class Node:
    def __init__(self, value): # T O(1)  S O(1) 
        self.value = value
        self.next = None
        
class LinkedList:
    def __init__(self, value): # T O(1)  S O(1)
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):  # T O(n)  S O(1)
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
        
    def append(self, value): # T O(1)  S O(1)
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1

    def pop(self):  # # T O(n)  S O(1)
        if self.length == 0:
            return None
        temp = self.head
        pre = self.head
        while(temp.next):
            pre = temp
            temp = temp.next
        self.tail = pre
        self.tail.next = None
        self.length -= 1
        if self.length == 0:
            self.head = None
            self.tail = None
        return temp

    def prepend(self, value): # T O(1)  S O(1)
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head
            self.head = new_node
        self.length += 1
        return True

    def pop_first(self): # T O(1)  S O(1)
        if self.length == 0:
            return None
        temp = self.head
        self.head = self.head.next
        temp.next = None
        self.length -= 1
        if self.length == 0:
            self.tail = None
        return temp

    def get(self, index): # T O(n)  S O(1)
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        for _ in range(index):
            temp = temp.next
        return temp
        
    def set_value(self, index, value): # T O(n)  S O(1)
        temp = self.get(index)
        if temp:
            temp.value = value
            return True
        return False
    
    def insert(self, index, value): # T O(n)  S O(1)
        if index < 0 or index > self.length:
            return False
        if index == 0:
            return self.prepend(value)
        if index == self.length:
            return self.append(value)
        new_node = Node(value)
        temp = self.get(index - 1)
        new_node.next = temp.next
        temp.next = new_node
        self.length += 1   
        return True  

    def remove(self, index): # T O(n)  S O(1)
        if index < 0 or index >= self.length:
            return None
        if index == 0:
            return self.pop_first()
        if index == self.length - 1:
            return self.pop()
        pre = self.get(index - 1)
        temp = pre.next
        pre.next = temp.next
        temp.next = None
        self.length -= 1
        return temp
        
    def reverse(self): # T O(n)  S O(1)
        temp=self.head
        self.head=self.tail
        self.tail=temp
        after=temp.next
        before=None
        for _ in range(self.length):
            after=temp.next
            temp.next=before
            before=temp
            temp=after
```

```python
# Leetcode Practice Q 

# 876. Middle of the Linked List (w/o using length)
class Solution:
    def middleNode(self, head: Optional[ListNode]) -> Optional[ListNode]:
        slow = fast = head 
        while fast is not None and fast .next is not None:
            slow=slow.next
            fast=fast.next.next 
        return slow
    
# 141. Linked List Cycle detection
class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        slow = fast =head 
        while fast is not None and fast.next is not None:
            slow=slow.next 
            fast=fast.next.next 

            if slow ==fast:
                return True 
        return False

# LC : Remove Nth Node From End of List (w/o using length)
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        dummy = ListNode(0, head)
        slow = fast = dummy
        for _ in range(n + 1):
            fast = fast.next
        while fast:
            slow = slow.next
            fast = fast.next
        slow.next = slow.next.next
        return dummy.next

# 1836. Remove Duplicates From an Unsorted Linked List
class Solution:
    def deleteDuplicatesUnsorted(self, head: ListNode) -> ListNode:
        freq_map = {}
        dummy = ListNode(-1, head)
        curr = head
        prev = dummy     

        while curr:
            freq_map[curr.val] = freq_map.get(curr.val, 0) + 1
            curr = curr.next

        curr = head
        while curr:
            if freq_map[curr.val] > 1:
                prev.next = curr.next
                curr = prev.next
            else:
                prev = curr
                curr = curr.next

        return dummy.next
'''
Input:  head = [3, 2, 2, 1, 3, 2, 4]
Counts: freq = {3:2, 2:3, 1:1, 4:1}
Output: [1, 4]
'''

# 1290. Convert Binary Number in a Linked List to Integer (w/o using length)
class Solution:
    def getDecimalValue(self, head: Optional[ListNode]) -> int:
        current = head
        result = 0
        while current:
            result = result * 2 + current.val
            current = current.next
        return result
    
# 86. Partition List (2 lists smaller and grater than x, then combine)
class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        if head is None:
            return None
        less_head = ListNode(0)
        greater_head = ListNode(0)
        less = less_head
        greater = greater_head
        current = head
        while current:
            if current.val < x:
                less.next = current
                less = current
            else:
                greater.next = current
                greater = current
            current = current.next
        greater.next = None

        less.next = greater_head.next

        return less_head.next
    
# 92. Reverse Linked List II 
class Solution:
    def reverseBetween(self, head: Optional[ListNode], left: int, right: int) -> Optional[ListNode]:
        if not head or left == right:
            return head
        
        # Step 1: Create a dummy node to simplify head operations
        dummy = ListNode(0)
        dummy.next = head
        prev = dummy

        # Step 2: Move prev to the node just before `left`
        for _ in range(left - 1):
            prev = prev.next
        
        # Step 3: Start reversing
        # `curr` is the first node of the sublist to reverse
        curr = prev.next
        for _ in range(right - left):
            temp = curr.next
            curr.next = temp.next
            temp.next = prev.next
            prev.next = temp
        
        return dummy.next
    
# 24. Swap Nodes in Pairs 
class Solution:
    def swapPairs(self, head: Optional[ListNode]) -> Optional[ListNode]:
        dummy = ListNode(0)
        dummy.next = head
        prev = dummy

        while head and head.next:
            first = head
            second = head.next

            # Swapping
            prev.next = second
            first.next = second.next
            second.next = first

            # Re-positioning pointers
            prev = first
            head = first.next

        return dummy.next
```
