```python
def printme(n):
    for i in range(n):
        print(i)

printme(10)

# Here Big of O in O(n)
```

```python
def printme(n):
    for i in range (n):
        for j in range (n):
            print(i,j) 

printme(10)

# here the Big O in O(n * n)
```

```python
def printme(n):
    for i in range (n):
        for j in range (n):
            print(i,j)

    for k in range(n):
        print(k) 



printme(10)

# This is called nested loops but still the Big O is O(n*n)
```

```python
def addme(n):
    print(n+n+n)

addme(10)

# This is called Big O of 1 i.e. O(1)
```

```python
def fun1(a):
    for i in range(a):
        print(a)

def fun2(b):
    for j in range(b):
        print(b)

# This is called "Different terms of inputs" and Big O is not O(n) but its O(a+b)
```
