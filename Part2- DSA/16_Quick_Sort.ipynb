```python
# Quick sort uses a pivot

'''
1] select a pivot
2] move the pivot to it's sorted position
3] Everything to left is small and right is big
4] Recursive fn on left and right side 

'''
```

```python
# Pivot a helper fn 

'''
1] Pick a pivot
2] pointer pivot, pointer i(index) just on right, pointer swap along pivot
3] i moves forward  

if i < pivot 
swap goes forward and swaps with i 

if i> pivot ; do nothing 

4] Swapping between swap and pivot 

result : all small on left big on right 
'''

def swap(my_list,index1,index2):
    temp = my_list[index1]
    my_list[index1] = my_list[index2]
    my_list[index2] = temp 
    
def pivot (my_list,pivot_index,end_index):
    swap_index= pivot_index

    for i in range(pivot_index + 1, end_index +1): # as for loop goes upto and and not including and so we have end_index + 1
        if my_list[i]<my_list[pivot_index]:
            swap_index += 1 
            swap(my_list,swap_index,i)
    swap(my_list,pivot_index,swap_index)
    return swap_index
```

```python
# Quick sort fm

def quick_sort_helper(my_list,left,right):

    if left < right:
        pivot_index = pivot(my_list,left,right)
        quick_sort_helper(my_list,left,pivot_index -1) 
        quick_sort_helper(my_list,pivot_index+1,right)
    return my_list

def quick_sort(my_list):
    return quick_sort_helper(my_list,0,len(my_list)-1)

```

```python
# Full Code   

def swap(my_list,index1,index2):   # O(n log n), O(n^2) for a sorted list   # T O(1)  S O(1)
    my_list[index1], my_list[index2] = my_list[index2], my_list[index1]
    
def pivot (my_list,pivot_index,end_index):   # T O(n)  S O(1)
    swap_index= pivot_index

    for i in range(pivot_index + 1, end_index +1): # as for loop goes upto and and not including and so we have end_index + 1
        if my_list[i]<my_list[pivot_index]:
            swap_index += 1 
            swap(my_list,swap_index,i)
    swap(my_list,pivot_index,swap_index)
    return swap_index

def quick_sort_helper(my_list,left,right):   # T O(n log n)  S O(n)

    if left < right:
        pivot_index = pivot(my_list,left,right)
        quick_sort_helper(my_list,left,pivot_index -1) 
        quick_sort_helper(my_list,pivot_index+1,right)
    return my_list

def quick_sort(my_list):   # T O(n^2)  S O(n)
    return quick_sort_helper(my_list,0,len(my_list)-1)
```

```python
# Quick sort with Hoare partition and Randomization

import random

def swap(my_list, index1, index2):   # T O(1)  S O(1)
    my_list[index1], my_list[index2] = my_list[index2], my_list[index1]

def hoare_partition(my_list, left, right):    # T O(n)  S O(1)
    # Randomized pivot selection
    pivot_index = random.randint(left, right)
    pivot_value = my_list[pivot_index]
    swap(my_list, pivot_index, left)  # Move pivot to start for simplicity

    i = left - 1
    j = right + 1

    while True:
        # Move i to the right until finding an element >= pivot
        i += 1
        while my_list[i] < pivot_value:
            i += 1

        # Move j to the left until finding an element <= pivot
        j -= 1
        while my_list[j] > pivot_value:
            j -= 1

        # If pointers cross, return j as the partition index
        if i >= j:
            return j

        # Swap elements that are on the wrong side
        swap(my_list, i, j)

def quick_sort_helper(my_list, left, right): # T O(n log n)  S O(n)
    if left < right:
        partition_index = hoare_partition(my_list, left, right)
        quick_sort_helper(my_list, left, partition_index)
        quick_sort_helper(my_list, partition_index + 1, right)
    return my_list

def quick_sort(my_list):  # T O(n log n)  S O(log n)
    return quick_sort_helper(my_list, 0, len(my_list) - 1)
```
